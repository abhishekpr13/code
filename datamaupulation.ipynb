```python
df = df.rename(columns={'wordcount':'Toatl_Word'})
df.head()
```

```python
#change data type
df['subj2_war_new'] = df['subj2_war'].astype(float)
df.head()
```

```python
df['subj2_war_new'] = df['subj2_war'].apply(lambda x: x+1)
df.head()
```

```python
## Data aggregating and grouping
df.head()
```

```python
groupedd_mean = df.groupby('subj2_war')['Toatl_Word'].mean()
print(groupedd_mean)
```

```python
grouped_sum = df.groupby(['id','sentiment_vol'])['Toatl_Word'].sum()
print(grouped_sum)
```

```python
##aggregate multiple function 
groupedd_agg = df.groupby('subj2_war')['Toatl_Word'].agg(['mean','sum','count'])
groupedd_agg
```

```python
#merging and joining the data frame 
#create a simplet data frame 
df1 = pd.DataFrame({'key':['A','B','C'],'value1':[1,2,3]})
df2 = pd.DataFrame({'key':['A','B','D'],'value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#Merge the data frame on key columns
pd.merge(df1,df2,on="key",how="inner")
```

```python
sp = pd.merge(df1,df2,on="key",how="outer")
print(sp)
```

```python
sp.fillna(0)
sp.rename(columns={'key':'Id'})
```
