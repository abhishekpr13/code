```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

#create a line plot 
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Basic line plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
#create a customized line lot

plt.plot(x,y,color = 'red',linestyle = '--',marker ='o',linewidth =2,markersize = 6)
plt.grid(True)
```

```python
#Multiple plots 
#sample data 

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y2,color = "red")
plt.title("plot-1")

plt.subplot(2,2,2)
plt.plot(x,y1,color = "blue")
plt.title("plot-2")

plt.subplot(2,2,3)
plt.plot(y2,x,color = "green")
plt.title("plot-3")

plt.subplot(2,2,4)
plt.plot(y1,y2,color = "orange")
plt.title("plot-4")
```

```python
#bar plot
categories = ['a','b','c','d','e']
values = [5,7,3,8,6]
#create bar plot
plt.bar(categories,values,color = "red")
plt.xlabel("categories")
plt.ylabel("values")
plt.title("categories bar graph")
plt.show()
```

```python
#sample data
data = [1,2,3,3,3,4,5,6,7,8,5,5,5,3,4]
#create a histogram
plt.hist(data,bins=5,color = 'orange',edgecolor ='black')
```

```python
#creating a scater plot
x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.scatter(x,y,color = "blue",marker='x')
```

```python
##pipe chart
lables = ['A','B','C','D']
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0.4) #move out from 1st and 4th slice

#creat a pie chart
plt.pie(sizes,explode=explode,labels=lables,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
#Pratical example
#sales data visulization 
import pandas as pd
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
##Plot total sales by product
sales_data_df.info()
```

```python
##plot total sales by product
total_sales_by_product = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
##plot sales trend over time
sales_trend = sales_data_df.groupby('Date')['Total Revenue' ].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
