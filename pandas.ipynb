```python
import pandas as pd
```

```python
#series
# It can any data type object
data = [1,2,3,4]
series = pd.Series(data)
print(series)
```

```python
#create a series for dictionary elements 
data = {"a":1,"b":2,"c":3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30,40]
index = ['a','b','c','d']
pd.Series(data,index)
```

```python
##data frame
##it is 2-dimensional data structured in row and column 
#create a data frame from dict of list
data = {
    'Name':['Abhishek','john','rahul'],
    'Age':[25,30,45],
    'city':['Boston','Dallas','New Delhi']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```
